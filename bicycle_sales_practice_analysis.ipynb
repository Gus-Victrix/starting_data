# Bike Store Sales

- In this notebook, I analyze the sales made on bike stores.

```python
# Availing relevant packages.
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as pyplot

# Setting matplotlib backend to inline
%matplotlib inline
```

## Loading data:

```python
sales = pd.read_csv('Data/sales_data.csv', parse_dates=['Date'])
```

```python
sales.head()
```

## What is the mean of `Customer_Age`?

```python
sales['Customer_Age'].mean()
```

### Plotting the density (KDE) and box plot of the `Cusetomer_Age` data.

```python
sales['Customer_Age'].plot(kind='kde', figsize=(14,6))
```

```python
sales['Customer_Age'].plot(kind='box', vert=False, figsize=(14,6))
```

## What's the mean `Order_Quantity`?

```python
sales['Order_Quantity'].mean()
```

```python
sales['Order_Quantity'].plot(kind='hist', bins=30, figsize=(14,6))
```

```python
sales['Order_Quantity'].plot(kind='box', vert=False, figsize=(14,6))
```

### Numper of sales per year

```python
sales['Year'].value_counts()
```

### Number of sales per year?

```python
sales['Year'].value_counts().plot(kind='pie', figsize=(7,7))
```

```python
sales['Month'].value_counts().plot(kind='pie', figsize=(8,8))
```
